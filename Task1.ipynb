```python
#计算缺失率和唯一率：删除一些不必要的列
data_1=pd.DataFrame(data.isnull().sum(),columns=['loss_num'])
data_2=pd.DataFrame(data.nunique(),columns=['unique_num'])
data_loss_unique=pd.merge(data_1,data_2,right_index=True,left_index=True)
data_loss_unique['loss_rate']=round(data_loss_unique['loss_num']/data.shape[0],5)
data_loss_unique['unique_rate']=round(data_loss_unique['unique_num']/data.shape[0],5)

data_loss_unique.head().T
```

```python
#查看缺失比较验证的列
data_loss_unique.sort_values(by='loss_rate',ascending=False)

#结合实际意义，可以将student_feature删掉
```

```python
#查看唯一值情况 唯一率特别高或者特别低的都可以考虑删除
data_loss_unique.sort_values(by='unique_rate',ascending=False)

#可以看出num、custid、id_name、source、trade_no、bank_card_no唯一率较高或者较低，且对于分类无实际意义，可直接删除
```

```python
#删除上述变量
columns_drop=['student_feature','bank_card_no','source','trade_no','id_name','num','custid']
data.drop(columns=columns_drop,axis=1,inplace=True)
```

```python
data.shape
```

```python
#obj对象需要进一步处理reg_preference_for_trad 、latest_query_time、loans_latest_time
```

```python
data['reg_preference_for_trad'].unique()
```

```python
data['latest_query_time'] = pd.to_datetime(data['latest_query_time'], format ='%Y-%m-%d')
data['loans_latest_time'] = pd.to_datetime(data['loans_latest_time'], format ='%Y-%m-%d')
```

```python
#处理reg_preference_for_trad，并将其转化为数值类型
for i,j in enumerate(data['reg_preference_for_trad']):
    if j=='一线城市':
        data['reg_preference_for_trad'][i]=1
    elif j=='二线城市':
        data['reg_preference_for_trad'][i]=2
    elif j=='三线城市':
        data['reg_preference_for_trad'][i]=3
    elif j=='其他城市':
        data['reg_preference_for_trad'][i]=4
    elif j=='境外':
        data['reg_preference_for_trad'][i]=5
    else:
        data['reg_preference_for_trad'][i]=0
```

```python
data['reg_preference_for_trad']=data['reg_preference_for_trad'].astype('float')
```

```python
#由于不是特别清楚loans_latest_time和latest_query_time的含义，计算新变量值为二者相减取绝对值，然后删除这两列
data['duration_time']=abs(data['latest_query_time']-data['loans_latest_time'])
for i,j in enumerate(data['duration_time']):
    data['duration_time'][i]=j.days
```

```python
data['duration_time']=data['duration_time'].astype('float')
```

```python
data.drop(['latest_query_time','loans_latest_time'],axis=1,inplace=True)
```
